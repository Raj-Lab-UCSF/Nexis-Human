# Generate figure for NEXIS humans

```python
import os
import numpy as np
import scipy as sp
import scipy.io
from scipy.linalg import expm
import pandas as pandas
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.optimize import minimize
from Nexis_human_functions import heatmap, normalize_by_l2_norm, mse_matrix, Nexis_error
from functools import partial
from scipy.stats import pearsonr
from Nexis_human_model import run_Nexis
# Added
from scipy.integrate import odeint, ODEintWarning
import warnings
```

```python
# List of sorted genes files excluding file results.csv
directory = 'genes'
genes_dir = os.listdir(directory)

genes_dir = [ele for ele in genes_dir if ele != 'results.csv']

genes_dir = sorted(genes_dir)

genes_names = genes_dir.copy()
for nn in range(100):
    genes_names[nn] = genes_dir[nn][:-4]
```

```python
# Loop over genes and calculate metrics
kk = 0
mean_beta_global = np.zeros(100,)
mean_beta_gene = np.zeros(100,)
mean_alpha_global = np.zeros(100,)
mean_alpha_gene = np.zeros(100,)
mean_gamma_global = np.zeros(100,)
mean_gamma_gene = np.zeros(100,)
mean_p_global = np.zeros(100,)
mean_p_gene = np.zeros(100,)
mean_b_global = np.zeros(100,)
mean_b_gene = np.zeros(100,)
mean_R_global = np.zeros(100,)
mean_R_gene = np.zeros(100,)

for file_name in genes_dir:
    file_path = os.path.join('genes', file_name)
    data = pandas.read_csv(file_path)
    
    mean_alpha_global[kk]=np.mean( data['alpha'][0::2] )
    mean_alpha_gene[kk]=np.mean( data['alpha'][1::2] )
    
    mean_beta_global[kk]=np.mean( data['beta'][0::2] )
    mean_beta_gene[kk]=np.mean( data['beta'][1::2] )
    
    mean_gamma_global[kk]=np.mean( data['alpha'][0::2] )
    mean_gamma_gene[kk]=np.mean( data['alpha'][1::2] )
    
    mean_p_global[kk]=np.mean( data['beta'][0::2] )
    mean_p_gene[kk]=np.mean( data['beta'][1::2] )
    
    mean_b_global[kk]=np.mean( data['alpha'][0::2] )
    mean_b_gene[kk]=np.mean( data['alpha'][1::2] )
    
    mean_R_global[kk]=np.mean( data['R squared'][0::2] )
    mean_R_gene[kk]=np.mean( data['R squared'][1::2] )
    
    kk=kk+1
```
